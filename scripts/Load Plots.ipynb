```python
root = "../CRISPResso_on_Linear_chr11_sm_201_left"
root = "../../tmp_test/CRISPResso_on_tmp"
crispresso2_info = CRISPRessoShared.load_crispresso_info(root)
```

```python
tot_reads = crispresso2_info['results']['alignment_stats']['counts_total'][ref_name]
```

```python
print(f'read {tot_reads} reads')
```

```python
nuc_pct_file = os.path.join(root,crispresso2_info['results']['refs'][ref_name]['nuc_pct_filename'])
mod_count_file = os.path.join(root,crispresso2_info['results']['refs'][ref_name]['mod_count_filename'])
```

```python
mod_count_file
```

```python
cols_in_file = list(pd.read_csv(nuc_pct_file, sep='\t', nrows=1,header=None ).values[0])
cols_in_file[0] = 'Nucleotide'
nuc_pct_df = pd.read_csv(nuc_pct_file,sep="\t",header=None,skiprows=[0])
nuc_pct_df.columns = cols_in_file
nuc_pct_df.insert(0,'Batch',ref_name)
```

```python
nuc_pct_df
```

```python
cols_in_file = list(pd.read_csv(mod_count_file, sep='\t', nrows=1,header=None ).values[0])
cols_in_file[0] = 'Modification'
mod_count_df = pd.read_csv(mod_count_file,sep="\t",header=None,skiprows=[0])
mod_count_df.columns = cols_in_file
first_col = mod_count_df.iloc[:,0]
last_row = mod_count_df.iloc[-1,:].values
mod_pct_df = mod_count_df.iloc[0:-1,1:].div(mod_count_df.iloc[-1,1:]) #divide by last row
mod_pct_df.insert(0,'Modification',first_col)
mod_pct_df.loc[len(mod_pct_df.index)] = last_row
mod_pct_df = mod_pct_df.copy()
mod_pct_df.insert(0,'Batch',ref_name)
```
